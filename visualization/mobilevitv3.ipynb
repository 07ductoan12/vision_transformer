```python
import json
import matplotlib.pyplot as plt
from typing import Dict, List
```

```python
with open("../experiments/mobilenetvit_flower/metrics.json") as f:
    metrics = json.load(f)
```

```python
def plot_loss_curves(results: Dict[str, List[float]]):
    
    # Get the loss values of the results dictionary (training and test)
    train_loss = results['train_loss']
    valid_loss = results['valid_loss']
    train_accuracy = results['train_acc']
    valid_accuracy = results['valid_acc']
    # Figure out how many epochs there were
    epochs = range(len(results['train_loss']))

    # Setup a plot 
    plt.figure(figsize=(23, 10))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_loss, label='train_loss')
    plt.plot(epochs, valid_loss, label='valid_loss')
    plt.scatter(epochs, train_loss, s=8)
    plt.scatter(epochs, valid_loss, s=8)
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accuracy, label='train_accuracy')
    plt.plot(epochs, valid_accuracy, label='valid_accuracy')
    plt.scatter(epochs, train_accuracy, s=8)
    plt.scatter(epochs, valid_accuracy, s=8)
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.legend()


plot_loss_curves(metrics)
```

```python
def plot_lr(results: Dict[str, List[float]]):
    
    # Get the loss values of the results dictionary (training and test)
    lr = results['lr']
    # Figure out how many epochs there were
    epochs = range(len(results['train_loss']))

    # Setup a plot 
    plt.figure(figsize=(23, 10))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, lr, label='Learning rate')
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.legend()
plot_lr(metrics)
```
